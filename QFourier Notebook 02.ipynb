## Quaternion circulant matrices, Quaternion Fourier matrices. Properties and applications
## Part II: Visualization of QFT matrices

```python
import numpy as np
from quaternion_circulant_matrix import create_qft_matrix
from quaternion_symplectic import get_modulus_phase_axis, vis
np.set_printoptions(precision=2, suppress=True)  # for compact output
```

```python
#########################################
#### Create a Quaternion Fourier Matrix
#########################################
E = create_qft_matrix(128, np.quaternion(0, 1, 0, 0))
_,_, axonas = get_modulus_phase_axis(E)
p = vis(E)
```

```python
E = create_qft_matrix(64, np.quaternion(0, 0, 1, 0))
p = vis(E) #, title='The QFT matrix with axis j')
```

```python
E = create_qft_matrix(256, np.quaternion(0, 0, 0, 1))
_,_, axonas = get_modulus_phase_axis(E)
p = vis(E) #, title='The QFT matrix with axis k')
```

```python
#################################################
# A QFT matrix with a randomly chosen axis mu
#################################################
axis = np.quaternion(0, np.random.randn(), np.random.randn(), np.random.randn()).normalized() # Axis must be a pure unit quaternion
E = create_qft_matrix(128, axis)
modu,phasi,axonas = get_modulus_phase_axis(E)
print(f'The QFT matrix with axis = {axis}')
p = vis(E) 
```

```python
## A QFT matrix Q, multiplied as pQp^{-1}, will yield another QFT matrix.
## p is a unit quaternion.
p = np.quaternion(np.random.randn(), np.random.randn(), np.random.randn(), np.random.randn()).normalized()
pinv = p.conjugate()
print(p*pinv) # This must be equal to 1 (possible bug for older numpy?)
D = p * E * pinv
vis(D)
```
