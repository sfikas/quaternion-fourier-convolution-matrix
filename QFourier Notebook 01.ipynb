## Quaternion circulant matrices, Quaternion Fourier matrices. Properties and applications
## Part I: Visualization of quaternion circulant matrices

```python
import numpy as np 
import quaternion
from quaternion_circulant_matrix import quaternionic_circulant_filter
from quaternion_symplectic import get_modulus_phase_axis, vis
np.set_printoptions(precision=2, suppress=True)  # for compact output
```

```python
#######################################
#### Create a toy circulant matrix
#######################################
## Start by creating a random matrix kernel
N = 100
kernel = quaternion.as_quat_array(np.random.randn(N, 4)*10)
C = quaternionic_circulant_filter(kernel)
print('The circulant matrix C is of size: {}'.format(C.shape))
print('Visualization of C (as modulus - phase - axis):')
p = vis(C)
#p.savefig('/tmp/circulant_random.png', bbox_inches = 'tight', pad_inches = 0)
```

```python
## A random *real* circulant matrix
# Only its 'magnitude' will be non-zero
N = 100
ttx = np.zeros([N, 4])
ttx[:, 0] = np.random.randn(N, )*10
kernel = quaternion.as_quat_array(ttx)
C = quaternionic_circulant_filter(kernel)
print('The circulant matrix C is of size: {}'.format(C.shape))
print('Visualization of C (as modulus - phase - axis):')
p = vis(C)
#p.savefig('/tmp/circulant_real.png', bbox_inches = 'tight', pad_inches = 0)
```

```python
## A random *complex* circulant matrix
N = 100
ttx = np.zeros([N, 4])
ttx[:, 0] = np.random.randn(N, )*10
ttx[:, 1] = np.random.randn(N, )*10
kernel = quaternion.as_quat_array(ttx)
C = quaternionic_circulant_filter(kernel)
print('The circulant matrix C is of size: {}'.format(C.shape))
print('Visualization of C (as modulus - phase - axis):')
modulo, phase, axonas = get_modulus_phase_axis(C)
p = vis(C)
#p.savefig('/tmp/circulant_complex.png', bbox_inches = 'tight', pad_inches = 0)
```
