# Proposition 3.1 & 3.2

### 3.1 The product Cx implements the quaternionic left convolution.
### 3.2 Quaternionic circulant matrices can be written as matrix polynomials.

```python
import numpy as np 
import quaternion
from quaternion_circulant_matrix import quaternionic_circulant_filter
from quaternion_symplectic import vis
np.set_printoptions(precision=2, suppress=True)  # for compact output
```

```python
################################
# Create a circulant matrix
################################
mykernel = np.array([
    np.quaternion(+0, -1, +1,  0),
    np.quaternion(+1, -2, -1, +2),
    np.quaternion(+2,  0, -2,  0),
    ]
)
# The implementation of the following functions acts a "constructive" argument for proposition 3.2.
# Also note, this implementation assumes (by convention) that the kernel becomes the first *column* of 
# the corresponding circulant matrix.
A = quaternionic_circulant_filter(mykernel)
print(A)
```

```python
########################################################################################################
# Corollary 3.2.1. Transpose and conjugate transpose are also circulant (with kernels as in paper)
########################################################################################################
print('****\n The transpose is also circulant, with kernel equal to {}:'.format(
    A.T[:, 0]
))
print(A.T)
print('****\n The conjugate transpose is also circulant, with kernel equal to {}:'.format(
    A.conj().T[:, 0]
))
print(A.conj().T)
```
